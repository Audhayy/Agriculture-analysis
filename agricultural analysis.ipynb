```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pyplot
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv(r'D:\agriculture\agri_dataset.csv')
```

```python
df.describe()
```

```python
df.head(15)
```

```python
df.head()
```

```python

laben = LabelEncoder()
df['Crop'] = laben.fit_transform(df['Crop'])

y=df['Crop']
y.head()
```

```python
x = df.drop('Yieldss ',axis = 1)
x =df.drop('State',axis=1)

x.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.6,random_state=46)
```

```python
dectre = DecisionTreeClassifier()
```

```python

dectre.fit(x_train,y_train)
```

```python
accuracy = dectre.score(x_train,y_train)
print('accuracy : {:.2f}'.format(accuracy*100))
```

```python
predict =dectre.predict(x_test)
print(predict)
```
